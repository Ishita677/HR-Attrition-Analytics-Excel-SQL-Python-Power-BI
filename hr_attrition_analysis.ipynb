# Section 5: Workforce Overview

### Insight

This reveals the total number of employees included in the dataset after data cleaning.


# Section 6: Attrition Analysis

## Attrition Count

```python
attrition_count = df['Attrition'].value_counts()

print("\nAttrition Count")
print(attrition_count)
```

### Insight

This shows how many employees stayed versus how many left the organization.

## Attrition Rate

```python
attrition_rate = (df['Attrition'] == 'Yes').mean() * 100

print("\nAttrition Rate:", round(attrition_rate,2), "%")
```

### Insight

The attrition rate indicates the percentage of employees who left the company.



# Section 7: Department Analysis

## Attrition by Department

```python
attrition_department = df[df['Attrition']=='Yes']['Department'].value_counts()

print("\nAttrition by Department")
print(attrition_department)
```

### Insight

Departments with higher attrition may experience workload imbalance or lower employee satisfaction.

# Section 8: Gender Analysis

## Attrition by Gender

```python
attrition_gender = df[df['Attrition']=='Yes']['Gender'].value_counts()

print("\nAttrition by Gender")
print(attrition_gender)
```

### Insight

This analysis identifies whether attrition patterns differ between genders.



# Section 9: Job Role Analysis

## Attrition by Job Role

```python
attrition_jobrole = df[df['Attrition']=='Yes']['JobRole'].value_counts()

print("\nAttrition by Job Role")
print(attrition_jobrole)
```

### Insight

Certain job roles may experience higher attrition due to workload, job stress, or lack of career growth opportunities.


# Section 10: Salary Analysis

## Average Salary by Department

```python
avg_salary = df.groupby("Department")["MonthlyIncome"].mean()

print("\nAverage Salary by Department")
print(avg_salary)
```

### Insight

This helps identify salary distribution across departments and whether lower salary groups correlate with higher attrition.

# Section 11: Attrition Visualization

## Attrition Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Distribution")
plt.show()
```

### Insight

This visualization shows the proportion of employees who stayed versus those who left.

## Attrition by Department Visualization


```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)

plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

### Insight

This chart highlights departments with higher employee turnover.

## Attrition by Gender Visualization

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Attrition', data=df)

plt.title("Attrition by Gender")
plt.show()
```

### Insight

This visualization compares attrition trends across genders.


## Attrition by Job Role Visualization

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', hue='Attrition', data=df)

plt.title("Attrition by Job Role")
plt.xticks(rotation=45)
plt.show()
```

### Insight

Roles with higher attrition can be identified, allowing HR teams to focus on retention strategies.

# Section 12: Salary vs Attrition

## Salary Comparison

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)

plt.title("Salary vs Attrition")
plt.show()
```

### Insight

Employees with lower salary ranges tend to leave more frequently.

# Section 13: Age Analysis

## Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)

plt.title("Employee Age Distribution")
plt.show()
```

### Insight

This distribution helps identify the dominant age groups within the organization.


# Section 14: Correlation Analysis

## Correlation Heatmap

```python
numeric_cols = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

### Insight

The heatmap shows relationships between numerical variables such as salary, experience, and job satisfaction.

# Section 15: Attrition Risk Roles

## Top 5 Job Roles with Highest Attrition

```python
top_attrition_roles = df[df['Attrition']=='Yes']['JobRole'].value_counts().head(5)

print("\nTop 5 Roles with Highest Attrition")
print(top_attrition_roles)
```

### Insight

These roles require targeted HR strategies to reduce employee turnover.

# Section 16: Work Environment Analysis

## Overtime vs Attrition

```python
overtime_attrition = pd.crosstab(df['OverTime'], df['Attrition'])

print("\nOvertime vs Attrition")
print(overtime_attrition)
```

### Insight

Employees working overtime show higher attrition due to possible burnout.


## Work Life Balance Impact

```python
worklife_attrition = pd.crosstab(df['WorkLifeBalance'], df['Attrition'])

print("\nWork Life Balance vs Attrition")
print(worklife_attrition)
```

### Insight

Lower work-life balance scores correlate with increased employee turnover.
